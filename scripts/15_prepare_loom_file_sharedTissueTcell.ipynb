```python
import scanpy as sc
import numpy as np
from scipy import sparse

adata = sc.read_h5ad(
    "/data/Blizard-AlazawiLab/rk/seurat/TissueExpandedTcells.h5ad"
)

# Use raw counts
if adata.raw is not None:
    adata.X = adata.raw.X

# Dense float32
if sparse.issparse(adata.X):
    adata.X = adata.X.toarray()

adata.X = adata.X.astype(np.float32)
adata.var_names_make_unique()

# Drop problematic Unicode obs columns
cols_to_drop = ["cell_type", "cell_type_with_cluster", "tissue.celltype"]
adata.obs = adata.obs.drop(
    columns=[c for c in cols_to_drop if c in adata.obs.columns]
)

# REQUIRED FOR PYSCENIC
adata.var["Gene"] = adata.var_names
adata.obs["CellID"] = adata.obs_names

# Write loom
adata.write_loom(
    "/data/Blizard-AlazawiLab/rk/seurat/TissueExpandedTcells.loom",
    write_obsm_varm=False
)

print("✅ Loom written with Gene + CellID")
```
